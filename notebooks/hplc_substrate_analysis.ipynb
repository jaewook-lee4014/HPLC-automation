# HPLC Substrate Consumption Analysis

Automated kinetic parameter extraction from HPLC time-series data.
Compares 1:1 and 2:1 Glucose:Xylose ratio experiments.

```python
import sys
sys.path.insert(0, '..')

import matplotlib
matplotlib.use('Agg')
import matplotlib.pyplot as plt
import pandas as pd

from hplc_analysis.pipeline import analyze_experiment, export_results
from hplc_analysis.pipeline import results_to_summary_df, results_to_timeseries_df
from hplc_analysis.plotting import (
    plot_consumption_curves,
    plot_consumption_rates,
    plot_normalized_consumption,
    plot_diauxic_summary,
    plot_cv_variability,
    plot_experiment_dashboard,
    plot_multi_experiment_comparison,
)

print('Imports OK')
```

## 1. Load Data

```python
DATA_DIR = '../data'

file_1to1 = f'{DATA_DIR}/1-to-1-GlucoseXyloseMicroplateGrowth (1).csv'
file_2to1 = f'{DATA_DIR}/2-to-1-GlucoseXyloseMicroplateGrowth (1).csv'

print('=== 1:1 Experiment ===')
res_1to1 = analyze_experiment(file_1to1)

print('\n=== 2:1 Experiment ===')
res_2to1 = analyze_experiment(file_2to1)

print('\nAnalysis complete.')
```

## 2. Raw Data Overview

```python
print('--- 1:1 Dataset ---')
print(f'Rows: {res_1to1.metadata.n_rows}, Timepoints: {res_1to1.metadata.n_timepoints}')
print(f'Time range: {res_1to1.metadata.time_range[0]:.1f} - {res_1to1.metadata.time_range[1]:.1f} h')
print(f'Substrates: {list(res_1to1.metadata.substrate_columns.keys())}')
print(f'Replicates per timepoint: {res_1to1.metadata.replicates_per_timepoint}')

print('\n--- 2:1 Dataset ---')
print(f'Rows: {res_2to1.metadata.n_rows}, Timepoints: {res_2to1.metadata.n_timepoints}')
print(f'Time range: {res_2to1.metadata.time_range[0]:.1f} - {res_2to1.metadata.time_range[1]:.1f} h')
print(f'Substrates: {list(res_2to1.metadata.substrate_columns.keys())}')
print(f'Replicates per timepoint: {res_2to1.metadata.replicates_per_timepoint}')
```

## 3. 1:1 Experiment — Kinetic Parameters

```python
print('=== 1:1 Kinetic Parameters ===')
for substrate, kin in res_1to1.kinetics.items():
    print(f'\n--- {substrate} ---')
    print(f'  S0 = {kin.S0:.3f} g/L')
    print(f'  Se = {kin.Se:.3f} g/L')
    print(f'  ΔS = {kin.delta_S:.3f} g/L')
    print(f'  Efficiency = {kin.efficiency:.1f}%')
    print(f'  q_avg = {kin.q_avg:.4f} g/L/h')
    print(f'  q_max = {kin.q_max:.4f} g/L/h')
    print(f'  t_qmax = {kin.t_qmax:.1f} h')
    print(f'  t_lag = {kin.t_lag:.1f} h')
    print(f'  t_depletion = {kin.t_depletion}')
    print(f'  t_50 = {kin.t_50}')
    print(f'  t_90 = {kin.t_90}')
    print(f'  t_active = {kin.t_active}')
    print(f'  q_active = {kin.q_active}')

print(f'\n--- Diauxic Analysis ---')
d = res_1to1.diauxic
print(f'  Detected: {d.diauxic_detected}')
print(f'  Primary: {d.primary_substrate}')
print(f'  Order: {d.consumption_order}')
print(f'  Diauxic lag: {d.t_diauxic_lag:.1f} h' if d.t_diauxic_lag else '  Diauxic lag: N/A')
print(f'  Overlap fraction: {d.overlap_fraction:.3f}' if d.overlap_fraction else '  Overlap: N/A')
```

```python
fig, ax = plt.subplots(figsize=(10, 5))
plot_cv_variability(res_1to1, ax=ax, title='1:1 Glucose:Xylose — Replicate CV%')
plt.tight_layout()
plt.savefig('../output/1to1_cv.png', dpi=150, bbox_inches='tight')
plt.show()
```

## 5. 2:1 Experiment — Kinetic Parameters

```python
print('=== 2:1 Kinetic Parameters ===')
for substrate, kin in res_2to1.kinetics.items():
    print(f'\n--- {substrate} ---')
    print(f'  S0 = {kin.S0:.3f} g/L')
    print(f'  Se = {kin.Se:.3f} g/L')
    print(f'  ΔS = {kin.delta_S:.3f} g/L')
    print(f'  Efficiency = {kin.efficiency:.1f}%')
    print(f'  q_avg = {kin.q_avg:.4f} g/L/h')
    print(f'  q_max = {kin.q_max:.4f} g/L/h')
    print(f'  t_qmax = {kin.t_qmax:.1f} h')
    print(f'  t_lag = {kin.t_lag:.1f} h')
    print(f'  t_depletion = {kin.t_depletion}')
    print(f'  t_50 = {kin.t_50}')
    print(f'  t_90 = {kin.t_90}')
    print(f'  t_active = {kin.t_active}')
    print(f'  q_active = {kin.q_active}')

print(f'\n--- Diauxic Analysis ---')
d = res_2to1.diauxic
print(f'  Detected: {d.diauxic_detected}')
print(f'  Primary: {d.primary_substrate}')
print(f'  Order: {d.consumption_order}')
print(f'  Diauxic lag: {d.t_diauxic_lag:.1f} h' if d.t_diauxic_lag else '  Diauxic lag: N/A')
print(f'  Overlap fraction: {d.overlap_fraction:.3f}' if d.overlap_fraction else '  Overlap: N/A')
```

## 8. Export Results to CSV

```python
OUTPUT_DIR = '../output'

s1, t1 = export_results(res_1to1, OUTPUT_DIR, '1to1_GlcXyl')
print(f'1:1 summary: {s1}')
print(f'1:1 timeseries: {t1}')

s2, t2 = export_results(res_2to1, OUTPUT_DIR, '2to1_GlcXyl')
print(f'2:1 summary: {s2}')
print(f'2:1 timeseries: {t2}')

# Combined summary
combined = pd.concat([
    results_to_summary_df(res_1to1, '1:1'),
    results_to_summary_df(res_2to1, '2:1'),
], ignore_index=True)
combined.to_csv(f'{OUTPUT_DIR}/batch_summary.csv', index=False)
print(f'\nBatch summary saved.')
print('\n', combined.T)
```

## 9. Quality Reports

```python
for label, res in [('1:1', res_1to1), ('2:1', res_2to1)]:
    print(f'\n=== {label} Quality Report ===')
    for substrate, qr in res.quality.items():
        print(f'\n  {substrate}:')
        print(f'    Max CV: {qr.max_cv:.1f}% at t={qr.max_cv_time:.1f}h')
        print(f'    High CV timepoints (>20%): {qr.high_cv_times}')
        print(f'    Monotonicity score: {qr.monotonicity_score:.2f}')
        if qr.increasing_intervals:
            print(f'    WARNING - Increasing intervals:')
            for t1, t2, delta in qr.increasing_intervals:
                print(f'      {t1:.1f} -> {t2:.1f}h: +{delta:.3f} g/L')

    # Outliers
    for substrate, ol in res.outliers.items():
        if len(ol) > 0:
            print(f'\n  {substrate} outliers:')
            print(ol.to_string(index=False))
```
